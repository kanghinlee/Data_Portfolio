```python
import seaborn as sns

# Simple Linear Regression Project
```

```python
iris = sns.load_dataset('iris')
iris
```

```python

iris = iris[['petal_length', 'petal_width']]
```

```python
x = iris['petal_length']
y = iris['petal_width']
```

```python
import matplotlib.pyplot as plt

# Scatter Plot between x (length) vs y (width)
plt.scatter(x,y)
plt.xlabel("petal_length")
plt.ylabel("petal_width")

# Note
# The plot shows position Correlation between Petal Length(X) and Petal Width(Y)
# Slight bias between 2 < x < 3, as there is no observes between the range
# Variant increase as Length increases (i.e. data is more spread out towards X > 5)
```

```python
# Split Train and Test Dataset

from sklearn.model_selection import train_test_split

# Keep 40% of testing data for test dataset (0.4)
# Keep a constrain to "random_state" to 42
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.4, random_state = 42)
```

```python
import numpy as np

# Convert series into array (i.e. 2 dimensional data)
x_train = np.array(x_train).reshape(-1,1)
x_test = np.array(x_test).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
c = lr.intercept_
#Intercept
c
```

```python
m = lr.coef_
# Gradient / x coefficient
m
```

```python
y_pre_train1 = lr.predict(x_train)
y_pre_train1
```

```python

# Scatter Plot between x (length) vs y (width)
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pre_train1, color = 'red')
plt.xlabel("petal_length")
plt.ylabel("petal_width")
```

```python
# validation using test model

# Y prediction using test data (X Test)
y_pre_test1 = lr.predict(x_test)

# Scatter Plot between x (length) vs y (width)
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pre_test1, color = 'red')
plt.xlabel("petal_length")
plt.ylabel("petal_width")
# The plot shows Position Correlation between X and Y with slight bias between 2 < x < 3
```
